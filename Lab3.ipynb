```r
library(tidyverse)
source("http://www.openintro.org/stat/data/cdc.R")
cdc <- as_tibble(cdc)
```

```r
str(cdc)
```

```r
ggplot(data = cdc) + 
  geom_bar(mapping = aes(x = genhlth, fill = gender), position="dodge")+
  xlab('General Health')+
  ylab('Count')+
  ggtitle('General Health for Males and Females')
```

```r
cdc$genhlth <- as.numeric(cdc$genhlth)
```

```r
# Health v height
ggplot(data = cdc, aes(x = height, y = genhlth, color = gender)) +  # Build a graph
  geom_point() +                                                     # Add a scatter plot
  ggtitle('General Health verses Height by Gender') +                         # Plot title
  xlab('Height') +                                              # x-label title
  ylab('General Health') +                                              # y-lable title
  scale_color_discrete(name = 'Gender',                              # Edit legend
                       labels = c('Male', 'Female')) +
  geom_smooth()
```

```r
genhlth_female=cdc$genhlth[cdc$gender == "f"]
print("Mean female genhlth: ")
mean(genhlth_female)

genhlth_male=cdc$genhlth[cdc$gender == "m"]
print("Mean male genhlth: ")
mean(genhlth_male)
```

```r
height_female=cdc$height[cdc$gender == "f"]
print("Mean female height: ")
mean(height_female)

height_male=cdc$height[cdc$gender == "m"]
print("Mean male height: ")
mean(height_male)
```

```r
female_frame=cdc[cdc$gender == 'f',]
```

```r
# Histogram of Femaile Height
g1<-ggplot(data =female_frame, aes(x = height,)) +  # Build a graph
  geom_histogram(bins = 20) +           # Add a histogram layer
  ggtitle('Histogram of Female Height') +        # Plot title
  xlab('Height') + # x-label title
  geom_vline(xintercept=64.367749976033) #Add vertical line for mean
```

```r
male_frame=cdc[cdc$gender == 'm',]
```

```r
# Histogram of Male Height
g2<-ggplot(data =male_frame, aes(x = height,)) +  # Build a graph
  geom_histogram(bins = 20) +           # Add a histogram layer
  ggtitle('Histogram of Male Height') +        # Plot title
  xlab('Height') + # x-label title
  geom_vline(xintercept=70.2516459400146) #Add vertical line for mean
```
